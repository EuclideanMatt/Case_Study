```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\Epicm\Desktop\data_bases\2U\data\2u_takehome_case.csv"
df = pd.read_csv(file_path)

df
```

```python
print(df.dtypes)
```

```python
df['id'] = range(1, len(df) + 1)

df
```

```python
df['spend'] = df['spend'].str.replace('$', '').str.replace(',','')

df
```

```python
# Parse 'month' column in 'Apr-20' format and transform to 'MM-YYYY'
df['month'] = pd.to_datetime(df['month'], format='%b-%y', errors='coerce').dt.strftime('%m-%Y')
df['quarter'] = pd.to_datetime(df['quarter'], format='%b-%y', errors='coerce').dt.strftime('%m-%Y')
df
```

```python
columns_to_convert = ['spend', 'leads', 'started_apps', 'predicted_submitted_apps']

# Convert the specified columns to floats
for col in columns_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce').astype(float)

df
```

```python
df['program'] = df['program'].str.replace(' ', '_')
df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)

df
```

```python
# Save the cleaned data to a new CSV (optional)
output_file = r"C:\Users\Epicm\Desktop\data_bases\2U\data\2u_takehome_case_cleaned.csv"
df.to_csv(output_file, index=False)
print(f"Cleaned data saved to {output_file}")
```

```python
# Calculate descriptive statistics for 'spend'
spend_stats = df['spend'].describe()

# Calculate the interquartile range (IQR)
Q1 = df['spend'].quantile(0.25)
Q3 = df['spend'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = max(0, Q1 - 1.5 * IQR)  # Adjusting to ensure no negative values
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['spend'] < lower_bound) | (df['spend'] > upper_bound)]

# Print summary
print("Descriptive Statistics for 'spend':")
print(spend_stats)
print(f"\nLower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Number of Outliers: {len(outliers)}")
```
